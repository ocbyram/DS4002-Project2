```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../DATA/weather.csv')
df.drop(['WESD', "WT01", "WT03", "WT04", "WT05", "WT06", "WT11"], axis=1, inplace=True)
df['date'] = pd.to_datetime(df['DATE'])
start_date = pd.to_datetime('2012-01-01')
df['days_since_2012'] = (df['date'] - start_date).dt.days
```

```python
import pandas as pd


# Convert DATE column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Extract year and month columns
df['YEAR'] = df['DATE'].dt.year
df['MONTH'] = df['DATE'].dt.month

# Calculate temperature range
df['TEMP_RANGE'] = df['TMAX'] - df['TMIN']

# Group by year and month
grouped = df.groupby(['YEAR', 'MONTH'])

# Calculate average temperature
avg_temp = grouped['TOBS'].mean()

# Calculate standard deviation of temperature
std_temp = grouped['TOBS'].std()

# Calculate highest high temperature
highest_high_temp = grouped['TMAX'].max()

# Calculate lowest low temperature
lowest_low_temp = grouped['TMIN'].min()

# Calculate largest change in temperature in two days
max_temp_change = grouped['TEMP_RANGE'].max()

# Print the results
print("Average Temperature (TOBS) in each month on a year by year basis:")
print(avg_temp)

print("\nStandard Deviation of Temperature (TOBS) in each month on a year by year basis:")
print(std_temp)

print("\nHighest High Temperature (TMAX) in each month on a year by year basis:")
print(highest_high_temp)

print("\nLowest Low Temperature (TMIN) in each month on a year by year basis:")
print(lowest_low_temp)

print("\nLargest Change in Temperature in two days in each month on a year by year basis:")
print(max_temp_change)



```

```python
from scipy.stats import pearsonr

# Filter out rows with missing values
df_cleaned = df.dropna(subset=['YEAR', 'TOBS', 'TMAX', 'TMIN', 'TEMP_RANGE'])

# Calculate correlation between year and weather values for each month
corr_avg_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TOBS'])[0])
p_value_avg_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TOBS'])[1])

corr_std_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TOBS'])[0])
p_value_std_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TOBS'])[1])

corr_highest_high_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TMAX'])[0])
p_value_highest_high_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TMAX'])[1])

corr_lowest_low_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TMIN'])[0])
p_value_lowest_low_temp = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TMIN'])[1])

corr_max_temp_change = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TEMP_RANGE'])[0])
p_value_max_temp_change = df_cleaned.groupby('MONTH').apply(lambda x: pearsonr(x['YEAR'], x['TEMP_RANGE'])[1])

# Print the correlation and p-value results
print("Correlation between Year and Average Temperature (TOBS) in each month:")
print(corr_avg_temp)
print("P-value for Average Temperature (TOBS) in each month:")
print(p_value_avg_temp)

print("\nCorrelation between Year and Standard Deviation of Temperature (TOBS) in each month:")
print(corr_std_temp)
print("P-value for Standard Deviation of Temperature (TOBS) in each month:")
print(p_value_std_temp)

print("\nCorrelation between Year and Highest High Temperature (TMAX) in each month:")
print(corr_highest_high_temp)
print("P-value for Highest High Temperature (TMAX) in each month:")
print(p_value_highest_high_temp)

print("\nCorrelation between Year and Lowest Low Temperature (TMIN) in each month:")
print(corr_lowest_low_temp)
print("P-value for Lowest Low Temperature (TMIN) in each month:")
print(p_value_lowest_low_temp)

print("\nCorrelation between Year and Largest Change in Temperature in two days in each month:")
print(corr_max_temp_change)
print("P-value for Largest Change in Temperature in two days in each month:")
print(p_value_max_temp_change)
```

```python
# Filter out correlation data based on p-values less than 0.05
significant_corr_avg_temp = corr_avg_temp[p_value_avg_temp < 0.05]
significant_corr_std_temp = corr_std_temp[p_value_std_temp < 0.05]
significant_corr_highest_high_temp = corr_highest_high_temp[p_value_highest_high_temp < 0.05]
significant_corr_lowest_low_temp = corr_lowest_low_temp[p_value_lowest_low_temp < 0.05]
significant_corr_max_temp_change = corr_max_temp_change[p_value_max_temp_change < 0.05]

# Print significant correlation results
print("Significant Correlation between Year and Average Temperature (TOBS) in each month:")
print(significant_corr_avg_temp)

print("\nSignificant Correlation between Year and Standard Deviation of Temperature (TOBS) in each month:")
print(significant_corr_std_temp)

print("\nSignificant Correlation between Year and Highest High Temperature (TMAX) in each month:")
print(significant_corr_highest_high_temp)

print("\nSignificant Correlation between Year and Lowest Low Temperature (TMIN) in each month:")
print(significant_corr_lowest_low_temp)

print("\nSignificant Correlation between Year and Largest Change in Temperature in two days in each month:")
print(significant_corr_max_temp_change)
```
